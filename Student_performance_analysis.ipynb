## **Libraries and Environment Setup**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## **Data Generation and Dataset Overview**

```python
np.random.seed(1)

data = {
    "Student_ID": range(1, 31),
    "Gender": np.random.choice(["Male","Female"], 30),
    "Study_Hours": np.random.randint(1, 10, 30),
    "Marks": np.random.randint(40, 95, 30)
}

df = pd.DataFrame(data)
df.head()
```

## **Dataset Structure and Summary Statistics**



```python
df.info()
df.describe()
```

## **Gender Distribution Analysis**

```python
df.groupby("Gender")["Student_ID"].agg(["count"])
```

## **Average Marks by Gender**

```python
df.groupby("Gender")["Marks"].mean().rename("Average Marks")
```

## **Relationship Between Study Hours and Marks**

```python
df.groupby("Study_Hours")["Marks"]\
.mean().rename("Average Marks")\
.sort_values(ascending=False)
```

## **Visualisation**

```python
plt.scatter(df["Study_Hours"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Relationship between Study Hours and Marks")
plt.show()
```
